```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# 1. Load data (Ensure the path matches your file location)
# If the CSV is in the same folder as this notebook, use 'yrbs_cleaned.csv'
# If you are running this from the 'notebooks' folder, use '../data/processed/yrbs_cleaned.csv'
df = pd.read_csv('yrbs_cleaned.csv') 

# 2. Display the first five rows to verify successful loading
print("--- Data Preview ---")
display(df.head())

# 3. Check the distribution of 'SadOrHopeless' (Initial EDA step)
print("\n--- Variable Statistics ---")
counts = df['SadOrHopeless'].value_counts()
print("Value Counts (1 = Yes, 2 = No):")
print(counts)

# 4. Check for missing values
missing_count = df['SadOrHopeless'].isnull().sum()
print(f"\nNumber of missing values: {missing_count}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style of the plots
sns.set_theme(style="whitegrid")

# 1. Create the bar chart for SadOrHopeless
# Added 'hue' and 'legend=False' to remove the FutureWarning
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='SadOrHopeless', data=df, hue='SadOrHopeless', palette='Set2', legend=False)

# 2. Add professional labels and title
plt.title('Distribution of Adolescent Depression Feelings\n(1 = Yes, 2 = No)', fontsize=14)
plt.xlabel('Response: Did you feel sad or hopeless?', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# 3. Add exact counts on top of each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

# 4. Display the plot
plt.show()
```
